```python
# Langkah 0 - Load Library
import pandas as pd 
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# Langkah 1 - Load Data
dpath = 'files/Titanic-Dataset-fixed.csv'
df = pd.read_csv(dpath)
df.head()
```

```python
# Langkah 2 - Slice Data
df = df[['Survived', 'Pclass', 'Age', 'Sex', 'Cabin']]
df.head()
```

```python
# Langkah 3 - Encoding
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex']) #proses encoding
df['Cabin'] = le.fit_transform(df['Cabin']) #proses encoding
```

```python
# Langkah 4 - Verifikasi Hasil 
df.head()
```

```python
# Langkah 5 - Standarisasi
std = StandardScaler()
df['Age'] = std.fit_transform(df[['Age']])
```

```python
df.head()
```

```python
df.to_csv('files/Titanic-Dataset-selected.csv', index=False)
```
